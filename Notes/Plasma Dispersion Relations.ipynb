# Unmagnetised Plasmas

## Electrostatic

### Langmuir Waves \ Electron Plasma Waves

$$
\omega^2 = \omega_p^2 + 3v_{th}^2 k^2
$$

with the caveat the $v_{ph} = \frac{\omega}{k} \ll v_{th}$

```python
# Plot frequency vs k
k = np.linspace(-2,2,200)
v_th = 0.5
w_p = 1
w = (w_p**2 + 3*v_th**2*k**2)**0.5
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.plot(k,w)
plt.title("Langmuir Wave Dispersion Relation")
plt.plot(k,3**0.5*v_th*np.abs(k),linestyle=":",label="$\omega=\sqrt{3}v_{th}$")
plt.axhline(w_p,label="Plasma Frequency",linestyle=":")
plt.axvline(0,color="black")
plt.ylim([0,np.max(w)])
plt.xlabel('k')
plt.ylabel('$\omega$')
plt.legend()

plt.subplot(122)
vp = w/k
vg = 3*v_th**2*k/w
plt.plot(k,vp,label="Phase Velocity")
plt.plot(k,vg,label="Group Velocity")
plt.title("Langmuir Wave Phase and Group Velocities")
plt.axhline(3**0.5*v_th,label="$\sqrt{3}v_{th}$",linestyle=":")
plt.axvline(0,color="black")
plt.ylim([-np.max(w),np.max(w)])
plt.xlabel('k')
plt.ylabel('v')
plt.legend()
```

### Ion-Acoustic

$$
\omega^2 = k^2 \frac{\gamma_iT_i}{m_i} + \frac{k^2\gamma_eT_e/m_i}{1+\gamma_ek^2\lambda_{De}^2}
$$

```python
# Plot frequency vs k
from astropy import constants
k = np.linspace(-10,10,200)
Te=1
Ti=0
ve = (Te/mr)**0.5
vi = Ti**0.5
mr = constants.m_e.value/constants.m_p.value
wi = 1
we = wi/mr**0.5
lam_de=ve/we

w = (k**2*vi**2 + k**2*Te/(1+k**2*lam_de**2))**0.5
plt.figure(figsize=(12,5))
plt.subplot(121)
plt.plot(k,w,label=f"Ti={Ti}")
plt.title("Ion-Acoustic Wave Dispersion Relation")
plt.axhline(we,label="Electron Plasma Frequency",linestyle=":")
plt.axhline(wi,label="Ion Plasma Frequency",linestyle=":")
plt.axvline(0,color="black")
plt.ylim([0,np.max(w)*1.5])
plt.xlabel('k')
plt.ylabel('$\omega$')
plt.legend()

plt.subplot(122)
vp = w/k
vg = 3*v_th**2*k/w
plt.plot(k,vp,label="Phase Velocity")
plt.plot(k,vg,label="Group Velocity")
plt.title("Ion Wave Phase and Group Velocities")
plt.axvline(0,color="black")
plt.ylim([-np.max(w),np.max(w)])
plt.xlabel('k')
plt.ylabel('v')
plt.legend()
```
