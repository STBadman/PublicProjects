## Successive Approximations of the Liouville Equation

### Somov - Plasma Astrophysics, Part I - Chapter 2/3

### Exact to Averaged Description

The exact distribution function for a collection of N point-like particles is :

$$ 
\hat{f}(\boldsymbol{r},\boldsymbol{v},t) = \sum_{i=1}^N \delta(\boldsymbol{r} - \boldsymbol{r}_i(t))\delta(\boldsymbol{v} - \boldsymbol{v}_i(t))
$$

Which is the phase space density as a function of position and velocity (6D phase space), i.e. how many particles there are per unit phase space density at position $(\boldsymbol{r},\boldsymbol{v})$ 

The exact distribution function obeys Liouville's equation :

$$
\bigg[ \frac{\partial}{\partial t} + \boldsymbol{v} \cdot \nabla_{\boldsymbol{r}} + \frac{\boldsymbol{\hat{F}}}{m} \cdot  \nabla_{\boldsymbol{v}} \bigg] \hat{f}(\boldsymbol{r},\boldsymbol{v},t) = \frac{D}{Dt} \hat{f}(\boldsymbol{r},\boldsymbol{v},t) =0
$$

under the assumption that $\nabla_{\boldsymbol{v}} \cdot \hat{\boldsymbol{F}} = 0 $ where $\boldsymbol{\hat{F}}(\boldsymbol{r},\boldsymbol{v},t)$ is the force felt by a particle at the phase space location $(\boldsymbol{r},\boldsymbol{v})$ at time $t$.

With the definition of the average over phase space volume $\Delta X = \Delta\boldsymbol{r}\Delta\boldsymbol{v}$, and time $\Delta t$:

$$
f(\boldsymbol{r},\boldsymbol{v}) = \frac{1}{\Delta X \Delta t}\int_{\Delta X, \Delta t} \hat{f}(\boldsymbol{r'},\boldsymbol{v'},t') d\boldsymbol{r}d\boldsymbol{v}dt
$$

after some algebra (integration by parts), and the definition of a mean force $\boldsymbol{\hat{F}} = \langle \boldsymbol{F} \rangle + \delta \boldsymbol{F}$, we can get the averaged Liouville Equation, Kinetic Equation or Collisional Boltzmann Equation (depending on what textbook you're reading!) :

$$
\bigg[ \frac{\partial}{\partial t} + \boldsymbol{v} \cdot \nabla_{\boldsymbol{r}} + \frac{\langle \boldsymbol{F} \rangle}{m} \cdot  \nabla_{\boldsymbol{v}} \bigg] f(\boldsymbol{r},\boldsymbol{v},t) = \bigg(\frac{\partial \hat{f}}{\partial t} \bigg)_c
$$

with the right hand signed (termed the collisional integral)

$$
\bigg(\frac{\partial \hat{f}}{\partial t} \bigg)_c = -\frac{1}{\Delta X \Delta t}\int_{\Delta X, \Delta t} \frac{1}{m} (\delta \boldsymbol{F} \cdot \nabla_{\boldsymbol{v}})\hat{f}d\boldsymbol{r}d\boldsymbol{v}dt
$$

i.e. the left hand side is the Liouville operator acting on a smoothed distibution function in a smooth force field, with all the non-smoothed effects confined to one term on the right. 

### Collisional Integral

Let's define the "spiky-ness" of the distribution function to be given by :

$$
\phi(\boldsymbol{r},\boldsymbol{v},t) =  \hat{f}(\boldsymbol{r},\boldsymbol{v},t) - f(\boldsymbol{r},\boldsymbol{v},t)
$$

where, by construction:

$$\frac{1}{\Delta X \Delta t}\int_{\Delta X, \Delta t}\phi(\boldsymbol{r},\boldsymbol{v},t)d\boldsymbol{r}d\boldsymbol{v}dt = \langle \phi(\boldsymbol{r},\boldsymbol{v},t) \rangle_{X,t} = 0 $$

and also by construction :

$$
\langle \boldsymbol{\delta F}(\boldsymbol{r},\boldsymbol{v},t) \rangle_{X,t} = 0 $$ 

This allows us to work on the collisional integral :

$$
\bigg(\frac{\partial \hat{f}}{\partial t} \bigg)_c = - \frac{1}{m} \langle(\delta  \boldsymbol{F} \cdot \nabla_{\boldsymbol{v}})\phi(\boldsymbol{r},\boldsymbol{v},t)\rangle_{X,t}
$$

We can work on the force term by considering it a superposition of 2 body interactions and also define the two point correlation function :

$$
f_{kl}(X,X_1,t) = \langle \phi(\boldsymbol{r},\boldsymbol{v},t) \phi(\boldsymbol{r_1},\boldsymbol{v_1},t) \rangle_t
$$

which eventually gives :

$$
\bigg(\frac{\partial \hat{f}_k}{\partial t} \bigg)_c = - \frac{1}{m_k}\nabla_{\boldsymbol{v}} \cdot \sum_l\int_{X_1}\boldsymbol{F}_{kl}(X,X_1) f_{kl}(X,X_1,t) dX_1
$$

where $F_{kl}(X,X_1)$ is the exact force exerted on a particle of species $k$ at position $X$ in phase space due to a particle of species $l$ at position $X_1$ in phase space. This holds for a general force without explicit time dependence, but an example is just the coulomb force :

$$
F_{kl}(X,X_1) = \frac{e_ke_l}{|\boldsymbol{r} - \boldsymbol{r_1} |^\frac{3}{2}}(\boldsymbol{r}-\boldsymbol{r_1})
$$

We see there is a linear contribution of all species (including self interaction) and the contribution from each species is weighted by the joint probability of the force-exerting species to be at any position $X_1$ given that the particle feeling the force is located at position $X$.

This join probability (correlation function) must be determined to make further progress. Somov makes progress by subtracting the exact Lioville Equation from the averaged representation, and eventually shows that to determine the two point correlation function, an equation involving the three point correlation function ($f_{klm} = \langle \phi_k(X_1,t)\phi_l(X_2,t)\phi_m(X_3,t)\rangle_t$) must be solved. This is therefore an unclosed system of equations.

### Zeroth Order : Vlasov Equation

To zeroth order, we neglect the right hand side entirely and obtain the Vlasov Equation for the smooth distribution function $f(\boldsymbol{r},\boldsymbol{v},t)$. If we study a plasma consisting of different type of particles indexed by the letter $k$, (charge $e_k$,mass $m_k$), experiencing electromagnetic forces only, with external current and charge densities $\boldsymbol{j^0},\rho^0$ and plasma current and charge densities $\boldsymbol{j^q},\rho^q$, the Vlasov Equation is:

$$
\bigg[ \frac{\partial}{\partial t} + \boldsymbol{v} \cdot \nabla_{\boldsymbol{r}} + \frac{e_k}{m_k}\bigg[\boldsymbol{E} + \boldsymbol{v} \times \boldsymbol{B} \bigg] \cdot  \nabla_{\boldsymbol{v}} \bigg] f(\boldsymbol{r},\boldsymbol{v},t) = 0
$$

With the electromagnetic field governed by Maxwell's Equations (Gaussian Units) :

$$
\nabla \cdot \boldsymbol{E} = 4 \pi (\rho^0 + \rho^q) \\
\nabla \times \boldsymbol{B} = \frac{1}{c}\partial_t \boldsymbol{E} + \frac{4 \pi}{c} (\boldsymbol{j^0} + \boldsymbol{j^q}) \\
\nabla \cdot \boldsymbol{B} = 0 \\
\nabla \times \boldsymbol{E} = -\frac{1}{c}\partial_t \boldsymbol{B}
$$

Note that the plasma charge and current is linear superposition of all plasma species individual components :

$$
\rho^q(\boldsymbol{r},t) = \sum_k e_k \int f_k(\boldsymbol{r},\boldsymbol{v},t) d\boldsymbol{v} \\ 
\boldsymbol{j^q}(\boldsymbol{r},t) = \sum_k e_k \int \boldsymbol{v} f_k(\boldsymbol{r},\boldsymbol{v},t) d\boldsymbol{v} \\ 
$$

and are moments of the smoothed distribution function and therefore themselves smoothed. 

The Vlasov Equation accounts for __collective effects__ only, but in spite of this is widely applicable to many plasma physics regimes in which the __interaction parameter__ is small. In particular gives rise to __linear waves__ and __landau damping__, among others.

A loose justification for interaction parameter deciding if the right hand side of the Kinetic Equation is seen by considering the rhs is essentially a coulomb force over a velocity scale times density fluctuations. The coulomb force is $e^2/l^2=e^2n^{2/3}$, the velocity scale is $v_s \sim k_BT$ and the density fluctuations are from poisson statistics $\sqrt{n}$. Multiplying these together goes as $e^2n^{1/3}/k_BT$ 

### First Order :  Landau Collisional Integral

The Landau collisional integral involves substituting the coulomb term for the general force in the above collisional integral and neglecting close Coulomb dollisions. This results in :

$$
\bigg(\frac{\partial \hat{f}_k}{\partial t} \bigg)_c = -\nabla_{\boldsymbol{v}} \cdot J_{k,\alpha}
$$

With 

$$
 J_{k,\alpha} = \frac{\pi e_k^2 \ln \Lambda_C}{m_k}\sum_l e_l^2 \int_\boldsymbol{v_l} \bigg\{f_k\frac{\partial f_l}{m_l\partial v_{l,\beta}}-f_l\frac{\partial f_k}{m_k\partial v_{k,\beta}}\bigg\}\frac{u^2\delta_{\alpha,\beta}-u_\alpha u_\beta}{u^3}d\boldsymbol{v}
$$

Where $\ln \Lambda_c = \ln \lambda_d/r_c$ is the coulomb logarithm (from integrating impact parameters) and $\boldsymbol{u} = \boldsymbol{v}-\boldsymbol{v_l}$.

You can see this is closed form and depends on the relative velocities integrated over velocity space.

### Second Order : Fokker-Planck Equation

Considering a spatially independent distribution function $f(\boldsymbol{v},t)$, the Fokker-Planck equation descibes it's evolution under overlapping weak collisions :

$$
\partial_t f = -\nabla_\boldsymbol{v} \cdot [\langle\delta \boldsymbol{v} \rangle f] + \nabla_\boldsymbol{v} \nabla_\boldsymbol{v} : [\langle \delta \boldsymbol{v} \delta \boldsymbol{v}\rangle f]  
$$

Where $\delta \boldsymbol{v}$ is the distibution of velocity changes due to these many weak collisions. Fokker-Planck is a general class of equations which describe the time evolution of a distribution of particles under the effects of damping (term 1) and diffusion (term 2).
