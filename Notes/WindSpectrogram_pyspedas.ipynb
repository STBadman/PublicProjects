```python
import numpy as np
from pyspedas import wind
from pytplot import get_data
from datetime import datetime,timedelta
```

```python
start = datetime(2021,1,1)
end = datetime(2021,1,2)
data = wind.waves(trange=[start,end])
```

```python
import matplotlib.pyplot as plt

R2 = get_data('E_VOLTAGE_RAD2')
R1 = get_data('E_VOLTAGE_RAD1')
TNR = get_data('E_VOLTAGE_TNR')

plt.pcolormesh([datetime.tR1[0],R1[2],R1[1].T,
               cmap="inferno",
               vmin = np.nanpercentile(R1[1].flatten(),5),
               vmax=np.nanpercentile(R1[1].flatten(),99))

plt.pcolormesh(R2[0],R2[2],R2[1].T,
               cmap="inferno",
               vmin = np.nanpercentile(R2[1].flatten(),5),
               vmax=np.nanpercentile(R2[1].flatten(),99))

plt.pcolormesh(TNR[0],TNR[2],TNR[1].T,
               cmap="inferno",
               vmin=np.nanpercentile(TNR[1].flatten(),5),
               vmax=np.nanpercentile(TNR[1].flatten(),99));

plt.yscale("log")

plt.show()
```
